```python
    #1

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 11)
y = 8 + 5 * x

plt.plot(x, y, color='blue', label='fare = 8 + 5 Ã— distance_km')

plt.title("Taxi fare vs Distance")

plt.xlabel("Distance (km)")
plt.ylabel("Fare (currency)")

plt.legend()

plt.xlim(0, 10)
plt.ylim(0, 60)

plt.savefig("taxi_fare_line.jpg", dpi=150, bbox_inches='tight')
plt.show()
```

```python
    #2.1

import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv("tips.csv")

# Create scatter plot: price_per_person (x) vs tip (y)
plt.figure(figsize=(5, 4))
plt.scatter(df['price_per_person'], df['tip'])

# Add title and axis labels
plt.title('Relationship between Price per Person and Tip', fontsize=14, fontweight='bold')
plt.xlabel('Price per Person ($)', fontsize=12)
plt.ylabel('Tip ($)', fontsize=12)

# Add grid for better readability
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
    #2.2

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")
max_per_day = df.groupby("day")["total_bill"].max()

plt.figure(figsize=(6, 4))
plt.bar(max_per_day.index, max_per_day.values)
plt.title("Max Total Bill Per Day")
plt.xlabel("Day")
plt.ylabel("Max Total Bill")
plt.tight_layout()
plt.show()
```

```python
    #2.3

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")

plt.figure(figsize=(6, 4))
plt.hist(df["tip"], bins=12, color="green")
plt.title("Histogram of Tips")
plt.xlabel("Tip Amount")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

    #Question: By looking at the graph- what is the range of the most common tip?   [2 - 3]
```

```python
    #2.4

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("tips.csv")
df["tip_perc"] = 100 * df["tip"] / df["total_bill"]

plt.figure(figsize=(6, 4))
plt.hist(df["tip_perc"], bins=12, color="purple")
plt.title("Histogram of Tip Percentage")
plt.xlabel("Tip Percentage (%)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

    #Question: By looking at the graph- what is the range of the most tip percentage?   [10 - 20]
```

```python
    #2.5

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("tips.csv")

fig, axes = plt.subplots(2, 2, figsize=(10, 7))

axes[0, 0].scatter(df['price_per_person'], df['tip'], color='blue')
axes[0, 0].set_title('Price per Person vs Tip')
axes[0, 0].set_xlabel('Price per Person ($)')
axes[0, 0].set_ylabel('Tip ($)')
axes[0, 0].grid(True, alpha=0.3)

max_bill_per_day = df.groupby('day')['total_bill'].max()
axes[0, 1].bar(max_bill_per_day.index, max_bill_per_day.values, color='orange')
axes[0, 1].set_title('Maximum Total Bill per Day')
axes[0, 1].set_xlabel('Day')
axes[0, 1].set_ylabel('Max Total Bill ($)')
axes[0, 1].grid(True, alpha=0.3, axis='y')

axes[1, 0].hist(df['tip'], bins=20, color='green', edgecolor='black')
axes[1, 0].set_title('Distribution of Tips')
axes[1, 0].set_xlabel('Tip ($)')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].grid(True, alpha=0.3, axis='y')

df['tip_perc'] = 100 * df['tip'] / df['total_bill']
axes[1, 1].hist(df['tip_perc'], bins=20, color='purple', edgecolor='black')
axes[1, 1].set_title('Distribution of Tip Percentage')
axes[1, 1].set_xlabel('Tip Percentage (%)')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].grid(True, alpha=0.3, axis='y')

fig.suptitle('Tips Dataset - Overview Analysis', fontweight='bold')

plt.tight_layout()

plt.savefig("tips_overview.png", dpi=300, bbox_inches='tight')

plt.show()
```
